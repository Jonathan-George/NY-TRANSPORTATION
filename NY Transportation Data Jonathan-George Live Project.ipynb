```python
import unittest
import pandas as pd
import numpy as np
```

```python
def calculate_correlation_matrix(df):
    return df.corr()

class TestCorrelationMatrix(unittest.TestCase):
    def test_correlation_matrix(self):
        # Create a sample DataFrame for testing
        data = {
            'A': np.random.rand(100),
            'B': np.random.rand(100),
            'C': np.random.rand(100),
            'D': np.random.rand(100)
        }
        df = pd.DataFrame(data)
```

```python
# Calculate the correlation matrix
correlation_matrix = calculate_correlation_matrix(df)

# Verify that the correlation matrix is symmetric
assert (correlation_matrix.values == correlation_matrix.T.values).all()

# Verify that the diagonal elements are all ones
assert (np.diag(correlation_matrix) == 1).all()

# Verify that correlation values are within the range [-1, 1]
assert (-1 <= correlation_matrix.values).all() and (correlation_matrix.values <= 1).all()

if __name__ == '__main__':
    # Load tests from the TestCorrelationMatrix class
    test_suite = unittest.TestLoader().loadTestsFromTestCase(TestCorrelationMatrix)

    # Create a test runner
    test_runner = unittest.TextTestRunner()

    # Run the tests
    test_runner.run(test_suite)
```
